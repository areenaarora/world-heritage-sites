```python
from plotnine import*
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none' # To make sure text transfers over properly
```

```python
df = pd.read_excel('whc-sites-2021.xls')
df.shape
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("-", "_")
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.groupby(by='states_name_en').count()
```

```python
(
    ggplot(df)
    + aes(x='region_en', y='danger')
    + geom_bar(size=4, stat="identity", color='#C9C7C1', fill='#C9C7C1', width=0.5)
    + coord_flip()
    + labs(title="Number of world heritage sites in danger by region")
    + theme_classic()
    + theme(figure_size=(10, 7))
)    
```

```python
df.groupby(by='region_en').plot()
```
